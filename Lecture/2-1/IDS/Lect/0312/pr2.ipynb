```python
import pandas as pd
import matplotlib as plt

df = pd.read_csv('assets/auto.csv')
df._get_numeric_data()
numeric = df.select_dtypes(include='number')
categorical = df.select_dtypes(exclude='number')
select_cols = ['year', 'displacement', 'weight']
new_data = df.loc[50:100, select_cols]

print("Numeric Columns:\n", numeric.head())
print("\nCategorical Columns:\n", categorical.head())
print("\nNew Data (Rows 50~100 with selected columns):\n", new_data.head())
```
