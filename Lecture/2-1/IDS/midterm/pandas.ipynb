```python
#실습5
df_titanic['isAlone'] = ((df_titanic['SibSp'] == 0) & (df_titanic['Parch'] == 0)).map({False : 0, True : 1})
df_titanic['Alone'] = ((df_titanic['SibSp'] == 0) & (df_titanic['Parch'] == 0)).map({False : 'No', True : 'Yes'})

df_titanic['class'] = df_titanic['Pclass'].map({3:'third',2:'second',1:'first'})
df_titanic['class'].head()
df_titanic['isAdultMale'] = ((df_titanic['Age'] > 18) & (df_titanic['Sex'] == 'male')).map({False : 0, True : 1})
df_titanic['isAdultMale'].head(10)
```

```python
#실습6
price_1 = df_titanic[df_titanic['Pclass'] == 1]['Fare'].mean()
price_2 = df_titanic[df_titanic['Pclass'] == 2]['Fare'].mean()
price_3 = df_titanic[df_titanic['Pclass'] == 3]['Fare'].mean()
print(f"1 : {price_1:.2f}, 2  : {price_2:.2f}, 3 : {price_3:.2f}")

svd_rate_MA = df_titanic[df_titanic['isAdultMale'] == 1]['Survived'].sum() / len(df_titanic['isAdultMale'] == 1) * 100
svd_rate_notMA = df_titanic[df_titanic['isAdultMale'] == 0]['Survived'].sum() / len(df_titanic['isAdultMale'] == 0) * 100
print(f"\nadult male survived rate : {svd_rate_MA:.2f}%")
print(f"not adult male survived rate : {svd_rate_notMA:.2f}%")

def div_age_3(age):
    if age <= 18:
        return 0
    if age > 60:
        return 2
    else: return 1

df_titanic['AgeGroup'] = pd.cut(df_titanic['Age'], bins=[0, 18, 60, float('inf')], labels=['Child/Teen','Adult','Elderly'])
surv = df_titanic.groupby('AgeGroup')['Survived'].mean()
surv.head()
```
