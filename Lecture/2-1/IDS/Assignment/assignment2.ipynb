## □ Income(수입) 분포를 성별(Gender)에 따라 나누어 시각화

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df = pd.read_csv('../assets/credit.csv')

sns.displot(data=df,x='Income',hue='Gender',kind='kde')
```

## □ Limit(신용 한도)의 분포를 결혼 여부(Married)에 따라 나누어 시각화

```python
sns.displot(data=df, x='Limit', hue='Married', element="step")
```

## □ Balance(잔고)의 분포를 인종(Ethnicity)별로 나누어 시각화

```python
sns.displot(data=df, x='Balance', hue='Ethnicity', kind='kde', fill=True)
```

## □ Married(결혼 여부)와 Student(학생 여부)에 따른 Balance(잔고)의 분포를 시각화

```python
sns.catplot(data=df, x='Married', y='Balance', kind='box', hue='Student')
```
