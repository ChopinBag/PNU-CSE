# Problem 4 - Prime Checker with Optimizations
Implement the function int is_optimized_prime(int n) that takes an integer n and returns 1 if n is prime and 0 otherwise.

You are given a skeleton code that checks if a number is prime. Your task is to optimize the prime-checking function using the following rules:

If a number ends in 0, 2, 4, 5, 6, or 8, then it's not prime (except for 2 and 5).
If the sum of the digits of a number is a multiple of 3, then the number is not prime (except for 3).
You should only check for divisibility up to the square root of the number.

```c
#include <stdio.h>
#include <math.h>

int is_optimized_prime(int n);

int main(void)
{
    int a;
    scanf("%d", &a);
    if (is_optimized_prime(a))
        printf("a prime\n");
    else
        printf("not a prime\n");
    return 0;
}

int is_optimized_prime(int n) 
{
    if (n == 2 || n == 5 || n == 3){
        return 1;
        }

    int nums[] = {0,2,4,5,6,8};
    for (int i=0;i<6;i++){
        if (n==nums[i]) {return 0;}
    }

    int digit = n;
    int sum = 0;
    while (digit!=0){
        sum += digit%10;
        digit /= 10;
    }
    if (sum%3==0){return 0;}
    int nn = (int)sqrt(n);
    
    for(int j=5;j<=nn;j++){
        if (n%j==0){return 0;}
    }

    return 1;
}
```
