## 학생 여부에 따른 수입과 잔고 관계: 
## 학생 여부에 따라 Income(수입)과 Balance(잔고) 사이의 관계를 시각화
### 그림사이즈: 10x6
### 제목: Income vs. Balance by Student Status and Gender
### Income을 x축, Balance를 y축으로 하는 relplot을 생성
### hue 옵션에 Student 열을 사용하여 학생과 비학생의 관계를 비교
### style 옵션을 Gender 열로 설정하여 남성과 여성의 스타일을 다르게 표시
### 학생 여부에 따라 수입과 잔고가 어떻게 달라지는지 해석해 보라

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

credits = pd.read_csv('assets/credit.csv')
sns.relplot(data=credits, x='Income', y='Balance', hue='Student', style='Gender')
plt.title('Incomde vs Balance by Student Status and Gender')
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 데이터 로드
data = pd.read_csv('assets/credit.csv')

# 시각화 테마 설정
sns.set_theme(style='darkgrid')

# 그림 크기 설정
plt.figure(figsize=(10, 6))

# relplot 생성
plot = sns.relplot(
    data=data,
    x='Income',
    y='Balance',
    hue='Student',
    style='Gender',
    kind='scatter'
)

# 제목 설정
plot.set(title='Income vs. Balance by Student Status and Gender')

# 그래프 표시
plt.show()
```

## 교육 수준에 따른 평균 잔고와 수입 관계 시각화: 
## 교육 수준(Education)에 따른 평균 잔고(Balance)와 수입(Income)의 관계

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

sns.relplot(data=credits, x='Income', y='Balance', hue = 'Gender', size='Cards')
plt.title('Incomde vs. Balance by Education')
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 데이터 로드
data = pd.read_csv('assets/credit.csv')

# 교육 수준과 성별로 그룹화하여 평균 Income, Balance, Cards 계산
avg_data = data.groupby(['Education', 'Gender']).agg({'Income': 'mean', 'Balance': 'mean', 'Cards': 'mean'}).reset_index()

# 시각화 테마 설정
sns.set_theme(style='darkgrid')

# 그림 크기 설정
plt.figure(figsize=(10, 6))

# 산점도 생성
plot = sns.scatterplot(
    data=avg_data,
    x='Income',
    y='Balance',
    hue='Gender',
    size='Cards',
    sizes=(20, 200)
)

# 제목 및 축 이름 설정
plt.title('Average Balance vs. Income by Education Level and Gender')
plt.xlabel('Average Income')
plt.ylabel('Average Balance')
plt.legend(title='Gender')

# 그래프 표시
plt.show()
```
