```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style='darkgrid')
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

## relplot

### 변수간의 관계 시각화에 초점 (특히 정량변수 간)

```python
palette = sns.color_palette('winter')
sns.relplot(data=tips, x='total_bill', y='tip', hue='size', style='time', size='size', col='time', palette=palette)
# hue : 추가적으로 구분할 범주형 변수
# style : 마커 표기등 스타일을 구분할 범주형 변수

plt.show()
```

```python
fmri = sns.load_dataset('fmri')
g = sns.relplot(x='timepoint', y='signal', kind='line', data=fmri, errorbar=None)
# kind='line' 을 통해 lineplot 으로 전환
```

## displot
- histplot
- kdeplot
- ecdfplot
- rugplot

### 정량 변수의 분포를 시각화

```python
df_pen = sns.load_dataset('penguins')
sns.displot(df_pen, x='flipper_length_mm', bins=20, hue='species', element='step')
#binwidth : 막대의 폭, bins : 막대 개수
```

```python
sns.displot(df_pen, y='flipper_length_mm', hue='species', multiple='stack')
```

```python
sns.displot(df_pen, x='flipper_length_mm', hue='species', stat='density', common_norm=False)
# stat = 'density' : 개수가 중요하지않고 밀도함수의 차이가 중요할 경우, 모두 표준화하여 확률밀도로 표현가능
```

```python
sns.displot(df_pen, x='bill_length_mm', y='bill_depth_mm',hue='species')
```

```python
sns.displot(df_pen, x='flipper_length_mm', kind='kde',row='species', 
            common_norm=False, height=2, aspect=4, fill=True, multiple='stack'
            )
# multiple : KDE 다양한 방법으로 표시가능
# fill : 밀도내부를 투명도 적용해 채울 수 있음
```

```python
sns.displot(df_pen, x='bill_length_mm', y='bill_depth_mm', hue='species', kind='kde')
```

```python
sns.jointplot(data=df_pen, x='bill_length_mm', y='bill_depth_mm')
# 기본값, scatterplot + histplot
```

```python
sns.jointplot(data=df_pen, x='bill_length_mm', y='bill_depth_mm', hue='species', kind='kde')
# kind='kde' 설정시 contour 그려짐
```

```python
g = sns.JointGrid(data=df_pen, x='bill_length_mm', y='bill_depth_mm')
g.plot_joint(sns.histplot)
g.plot_marginals(sns.boxplot)
# 그래프 지정가능
```

## 실습
### 학생 여부에 따른 수입과 잔고 관계: 학생 여부에 따라 Income(수입)과 Balance(잔고) 사이의 관계를 시각화

- 그림사이즈: 10x6
- 제목: Income vs. Balance by Student Status and Gender
- Income을 x축, Balance를 y축으로 하는 relplot을 생성
- hue 옵션에 Student 열을 사용하여 학생과 비학생의 관계를 비교
- style 옵션을 Gender 열로 설정하여 남성과 여성의 스타일을 다르게 표시
- 학생 여부에 따라 수입과 잔고가 어떻게 달라지는지 해석해 보라

```python
df_credit = pd.read_csv('../assets/credit.csv')
g = sns.relplot(data=df_credit,x='Income', y='Balance', hue='Student', style='Gender', height=10, aspect=10/6)
g.figure.suptitle('Income vs. Balance by Student Status and Gender', y=1.03)
```

## 실습
### 교육 수준에 따른 평균 잔고와 수입 관계 시각화: 교육 수준(Education)에 따른 평균 잔고(Balance)와 수입(Income)의 관계 
- 각 Education(교육 수준)과 Gender(성별)에 대해 평균 Income(수입)과 평균 Balance(잔고), 평균 카드 개수를 계산
- relplot을 사용하여 Income을 x축, Balance를 y축으로 하여 산점도를 생성
- hue 옵션에 Gender를 추가하여 성별 차이를 시각화하고, size 옵션을 Cards로 설정하여 카드 개수에 따라 점 크기를 다르게 표시
- 그래프 제목과 축 이름을 설정
- 교육 수준과 성별에 따른 평균 잔고와 수입의 관계를 해석

```python
g2 = sns.relplot(data = df_credit, x='Income', y='Balance', hue='Gender', size='Cards')
g2.figure.suptitle('Balance vs Income by Education')
```

```python
sns.set_style("whitegrid")

g2 = sns.relplot(
    data=df_credit,
    x='Income',
    y='Balance',
    hue='Gender',         # 색상: 성별
    style='Education',    # 모양: 교육 수준
    size='Cards',         # 크기: 카드 개수
    sizes=(50, 300),      # 최소·최대 점 크기
    markers=True,         # education 별 기본 마커 사용
    alpha=0.8,            # 반투명도 조절
    legend='full',
    height=10
)

# 제목 및 축 레이블 설정

g2.figure.suptitle('Balance vs Income by Education, Gender and Number of Cards', y=1.02)
g2.set_axis_labels('Income', 'Balance')

plt.show()
```

## Income(수입) 분포를 성별(Gender)에 따라 나누어 시각화
- seaborn의 displot을 사용하여 Income의 분포를 시각화
- hue 옵션에 Gender를 추가하여 성별에 따라 구분된 분포를 표시
- kind 옵션을 kde로 설정하여 커널 밀도 추정(KDE) 그래프 그리시오

```python
sns.displot(data=df_credit, x='Income', hue='Gender', kind='kde')
```

## Limit(신용 한도)의 분포를 결혼 여부(Married)에 따라 나누어 시각화
- displot을 사용하여 Limit의 분포를 히스토그램 형태로 시각화
- hue 옵션에 Married를 추가하여 결혼 여부에 따라 색상으로 구분된 분포를 표시
- element="step" 옵션을 사용해 분포를 누적 선형 스타일로 표시

```python
sns.displot(data = df_credit, x='Limit', hue='Married', element='step')
```

## Balance(잔고)의 분포를 인종(Ethnicity)별로 나누어 시각화
- displot을 사용하여 Balance 열의 분포를 시각화
- hue 옵션에 Ethnicity를 추가하여 인종별로 구분된 분포를 나타내라
- kind="kde"를 사용하여 KDE 곡선을 표시
- fill=True 옵션을 추가하여 각 KDE 곡선을 채워서 표시

```python
sns.displot(data=df_credit, x='Balance', hue='Ethnicity', kind='kde', fill=True)
```

## Married(결혼 여부)와 Student(학생 여부)에 따른 Balance(잔고)의 분포를 시각화
- seaborn의 catplot을 사용하여 Married를 x축에, Balance를 y축에 두고 상자 그림(kind="box")을 생성
- hue 옵션에 Student를 추가하여 학생 여부에 따라 색상으로 구분된 상자 그림을 표시

```python
sns.catplot(data = df_credit, x='Married', y='Balance', kind='box', hue='Student')
```
