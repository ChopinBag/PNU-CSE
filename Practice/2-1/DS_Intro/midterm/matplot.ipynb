# Matplot

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('../assets/auto.csv')
plt.hist(df['horsepower'], bins=20) # bins 로 구간의 개수 조정
```

```python
plt.scatter(df['weight'], df['mpg'])
```

```python
plt.scatter(df['cylinders'], df['weight'])
```

```python
pies = df.groupby('cylinders')['horsepower'].count()
plt.pie(pies, labels=pies.index)
```

```python
bars = df.groupby('origin')['weight'].mean()
plt.bar(x=bars.index, height=bars)
```

```python
df_bike = pd.read_csv('../assets/bike.csv')
plt.plot(df_bike['temp'])
```

```python
plt.boxplot(df_bike['cnt'])
```

## 실습1

```python
df_bike['dteday'] = pd.to_datetime(df_bike['dteday'])
monthly = df_bike.groupby(df_bike['dteday'].dt.to_period("M"))['cnt'].sum()
monthly.plot()
plt.show()
```
