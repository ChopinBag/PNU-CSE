## 연습문제 1

```python
cm = df_bike.groupby('hr')['casual'].mean()
rm = df_bike.groupby('hr')['registered'].mean()

plt.title('Average Bike Rentals by Hour')
plt.plot(cm,color='blue', marker='o',)
plt.plot(rm,color='orange', marker='x',)

plt.xlabel('Hour of a Day')
plt.ylabel('Average Rentals')

ymax = max(cm.max(), rm.max())

plt.xticks([0,23])
plt.yticks((np.linspace(0, ymax, 6))) # 어떻게 처리해야하지

plt.grid(axis='both', linewidth=0.5)

plt.legend()
plt.show()
```

## 연습문제 2

```python
sc = df_bike.groupby('season')['cnt'].sum()
cls = ['blue', 'green', 'red', 'gray']
plt.figure(figsize=(8,6))

plt.bar(x=sc.index, height=sc, width=0.6, color=cls)
plt.xticks([1,2,3,4],['spring', 'summer', 'fall', 'winter'])
plt.xlabel('season')
plt.ylabel('total rentals')
plt.title('total bike rental by season')

plt.show()
```

## 연습문제 3

- labels: 계절을 나타내는 레이블 (Spring, Summer, Fall, Winter)을 설정
- 각 파이 조각의 비율을 소수점 첫째 자리까지 퍼센트로 표시
- 파이 차트가 90도에서 시작되도록 설정
- 색상: 계절에 따라 각각 다른 색상(red, green, blue, grey)
- 첫 번째 조각(Spring)을 강조하기 위해 파이 차트에서 약간 분리
- 제목: Bike Rentals by Season 으로 하고 bold 처리


```python
seasons = ['spring','summer','fall','winter']
colors2 = ['red','green','blue','grey']
expl = [0.1 ,0,0,0]
plt.pie(x=sc, autopct='%.1f%%', labels=seasons, startangle=90, colors=colors, explode=expl)
plt.title('Bike Rentals by Season', fontweight='bold')

plt.show()
```

## 연습문제 4
### 시간대에 따른 총 청구 금액(total_bill)과 팁(tip)의 관계를 분석하고, 이를 산점도로 시각화
- 시간대(time)을 기준으로 데이터를 Lunch와 Dinner 두 그룹
- x축에는 총 청구 금액(total_bill), y축에는 팁(tip)을 둠
- Lunch 데이터는 파란색(blue) 점으로, Dinner 데이터는 주황색(orange) 점
- alpha=0.6, markersize=10
- 제목: Total Bill vs Tip by Time 으로 bold 처리
- x축/y축 레이블: Total Bill / Tip
- 범례추가
- 그리드: linestyle='--', color='gray', alpha=0.7

```python
df_tips = pd.read_csv('../assets/tips.csv')

lunch = df_tips[df_tips['time'] == 'Lunch']
dinner = df_tips[df_tips['time'] == 'Dinner']

plt.scatter(x = lunch['total_bill'], y= lunch['tip'], color='blue', alpha=0.6, s=10, label='lunch')
plt.scatter(x = dinner['total_bill'], y= dinner['tip'], color='orange', alpha=0.6, s=10, label = 'dinner')

plt.title('Total bill vs Tip by Time',fontweight='bold')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.legend()
plt.grid(linestyle='--', color='gray', alpha=0.7)

plt.show()
```
