# Matplot

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
df = pd.read_csv('../assets/auto.csv')
plt.hist(df['horsepower'], bins=20) # bins 로 구간의 개수 조정
```

```python
plt.scatter(df['weight'], df['mpg'])
```

```python
plt.scatter(df['cylinders'], df['weight'])
```

```python
pies = df.groupby('cylinders')['horsepower'].count()
plt.pie(pies, labels=pies.index)
```

```python
bars = df.groupby('origin')['weight'].mean()
plt.bar(x=bars.index, height=bars)
```

```python
df_bike = pd.read_csv('../assets/bike.csv')
plt.plot(df_bike['temp'])
```

```python
plt.boxplot(df_bike['cnt'])
```

## 실습1

```python
df_bike['dteday'] = pd.to_datetime(df_bike['dteday'])
monthly = df_bike.groupby(df_bike['dteday'].dt.to_period("M"))['cnt'].sum()
monthly.plot()
plt.show()
```

```python
plt.figure(figsize=(12, 6))

plt.plot(df_bike['temp'], label='temp', color='lightgreen')
plt.plot(df_bike['atemp'], label='atemp', color='lightblue')

plt.xlabel('Days')
plt.ylabel('Temperature')

plt.title('Days vs Temperature')

plt.xticks(np.arange(0, 17500, 5000))
plt.yticks(np.arange(0,1,0.1))

plt.grid()

plt.xlim(0, 18000)
plt.ylim(0,1)

plt.legend()
plt.show()
```

```python
x=[1,2,3,4]
y=[2,3,5,10]
y2=[1,2,4,8]
plt.plot(x, y, 'bo--', marker='s')
plt.plot(x,y2, marker='^', markersize=9)

plt.xlabel('X-Asis')
plt.ylabel('Y-Asis')

plt.fill_between(x[1:3], y[1:3],y2[1:3] ,alpha=0.5)
plt.fill_betweenx(y[2:4], x[2:4], alpha=0.3, color='black')

plt.grid(True, axis='x', linestyle='--')

plt.axhline(4.0, 0.1, 0.5, color='lightgray', linestyle='--', linewidth=2)
plt.axvline(0.34, 0.1, 0.5, color='green', linestyle='--', linewidth=2)

plt.hlines(6, 1.0, 2.5, color='red',linestyle='solid',linewidth=3)
plt.vlines(0.59,2.0,8.7, color='skyblue',linewidth=4)

plt.legend()
plt.show()
```

```python
x = [1,2,3,4]
y=[0,3,4,9]
yerr=[1.1, 2.3, 1.5, 0.5]

plt.errorbar(x,y,yerr=yerr)
plt.show()
```
