```python
import matplotlib.pyplot as plt
import pandas as pd

df_auto = pd.read_csv('assets/auto.csv')

plt.hist(df_auto['horsepower'], bins=20)
```

```python
pies = df_auto.groupby('cylinders')['horsepower'].count()
plt.pie(pies, labels=pies.index)
```

```python
plt.plot(df_auto['acceleration'])
```

```python
plt.boxplot(df_auto['acceleration'])
```
