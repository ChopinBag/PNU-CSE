```python
import pandas as pd
import matplotlib.pyplot as plt

plt.plot([1,2,3,4], [2,3,5,10], 'bo-')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.show()
```

```python
import numpy as np

plt.grid(True, axis='y', color='red', alpha=0.5, linestyle = '--')
plt.grid(True, axis='x', color='blue', alpha=0.5, linestyle = '--')

x = [1,2,3,4]
y1 = [2,3,5,10]
y2 = [1,2,4,8]

plt.plot(x,y1, marker='^')
plt.plot(x,y2, marker='o')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.fill_between(x[1:3],y1[1:3],y2[1:3],color='gray',alpha=0.5)
plt.show()
```
