```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import scipy.stats as stats
```

```python
dof = 2;
x=np.arange(-4, 4, 0.01)
y=stats.t.pdf(x,dof)

plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('t distribution (dof=2)')
plt.show()
```

```python
y1 = stats.t.cdf(2, dof)
y2 = stats.t.cdf(-2,dof)

print(f'p(-2 <= t <= 2)= {y1-y2:.2f}')
```

```python
percent_point = 0.95
print(f'P(x<=k)=0.95, k = {stats.t.ppf(percent_point,dof):.2f}')
```
